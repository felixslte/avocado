## Explorative Datenanalyse





### Welches Produkt ist das beliebteste (gemessen an den Absatzzahlen Total US)?
1. Organic & Conventional


```python
total_US = df[df['region']=='TotalUS']
```

```python
fig = px.scatter(total_US, x="Date", y="AveragePrice", color="type",title="Preisentwicklung fr TotalUS nach Typ")
fig.show()
```

Was sieht man?
- organic 2015 - Mitte 2016 deutlich teuerer im Durchschnitt, klare Abgrenzung
- Ende 2016 - Anfang 2017 Grenzen nicht mehr so deutlich
- annhernd gleichgerichtete Auf und Abbewegungen

- Frage: Was ist mit Daten im Juli/August 2015? 

```python
fig = px.scatter(total_US, x="Date", y="Total Volume", color="type",title="Absatzzahlen - Organic & Conventional im Vergleich")
fig.show()
```

```python
fig = px.scatter(total_US[total_US['type']=="organic"], x="Date", y="Total Volume",color='type',title="Absatzzahlen organic")
fig.show()
```

- Absatzzahlen organic: positiver Trend ber die Jahre hinweg
- --> Interpretation: Nachhaltigkeitsbewusstsein in der Bevlkerung wchst

### Regression Preis & Absatzzahlen

```python
fig = px.scatter(total_US, x="AveragePrice", y="Total Volume", color="type",trendline='ols',title="Lineare Regression Absatzzahlen & Preis")
fig.show()
```

```python
fig = px.scatter(total_US[total_US['type']=="organic"], x="AveragePrice", y="Total Volume",color='type', trendline='ols',title="Korrelation Preis & Absatzzahlen organic")
fig.show()
```

Contional:
- deutliche negative Korrelation 
- mit steigendem Preis sinken die Abstze

Organic:
- mit steigendem Preis bleiben Absatzzahlen ungefhr konstant
- Warum? Organic Kunden sind berzeugt von Qualitt, bereit dafr mehr auszugeben

### 2. Bags & size

```python
total_US_conventional = total_US[total_US['type']=="conventional"]
total_US_organic = total_US[total_US['type']=="organic"]

fig = px.scatter(total_US_conventional, x="Date", y=["Small Bags","Large Bags","XLarge Bags"],title="Abzatzzahlen der Bags - conventional")
fig.show()
```

- Small Bags am beliebtesten

```python
fig = px.scatter(total_US_organic, x="Date", y=["Small Bags","Large Bags","XLarge Bags"],title="Abzatzzahlen der Bags - organic")
fig.show()
```

- Unterschied zwische Small & Large nicht so gro wie bei conventional
- XLarge gibt es nicht
- ab Mitte Juli 2017 sehr deutliche berlegenheit der Small Bags

```python
fig = px.scatter(total_US_conventional, x="Date", y=["4046","4225","4770"],title="Abzatzzahlen der Sorten (Gre) - conventional")
fig.show()
```

- 4046, 4225 hnliche Verkaufszahlen
- 4770 sehr geringe Absatzzahlen (extra large Hass), wahrscheinlich auch geringeres Angebot 

```python
fig = px.scatter(total_US_organic, x="Date", y=["4046","4225","4770"],title="Abzatzzahlen der Sorten (Gre) - organic")
fig.show()
```

- 4225 deutlich mehr verkauft als 4046
- 4770 hnlich geringe Absatzzahlen

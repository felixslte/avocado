```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime as date
```

```python
avocado_df = pd.read_csv('Avocado.csv')
```

```python
avocado_df.head(100)
```

```python
avocado_df.plot(x="region")
```

```python
regions1 = avocado_df["region"].to_numpy() #get a list of regions
regions=np.unique(regions1)
print(regions)
```

```python
#get a list of jahreszeiten, Jahreszeiten von Wikipedia genommen
dates1 = avocado_df["Date"].to_numpy()     #get a list of dates as datedatatype                                     
dates1asdates=[]
for i in range(0,dates1.size):
    dates1asdates.append(date.strptime(dates1[i],"%Y-%m-%d"))
#Jahreszeiten abgespeichert
seasons2015 = [('winter', (date(2015,  1,  1),  date(2015,  3, 19))),
           ('spring', (date(2015,  3, 20),  date(2015,  6, 20))),
           ('summer', (date(2015,  6, 21),  date(2015,  9, 22))),
           ('autumn', (date(2015,  9, 23),  date(2015, 12, 21))),
           ('winter', (date(2015, 12, 22),  date(2015, 12, 31)))]
seasons2016 = [('winter', (date(2016,  1,  1),  date(2016,  3, 19))),
           ('spring', (date(2016,  3, 20),  date(2016,  6, 20))),
           ('summer', (date(2016,  6, 21),  date(2016,  9, 21))),
           ('autumn', (date(2016,  9, 22),  date(2016, 12, 20))),
           ('winter', (date(2016, 12, 21),  date(2016, 12, 31)))]
seasons2017 = [('winter', (date(2017,  1,  1),  date(2017,  3, 19))),
           ('spring', (date(2017,  3, 20),  date(2017,  6, 20))),
           ('summer', (date(2017,  6, 21),  date(2017,  9, 21))),
           ('autumn', (date(2017,  9, 22),  date(2017, 12, 20))),
           ('winter', (date(2017, 12, 21),  date(2017, 12, 31)))]
seasons2018 = [('winter', (date(2018,  1,  1),  date(2018,  3, 19))),
           ('spring', (date(2018,  3, 20),  date(2018,  6, 20))),
           ('summer', (date(2018,  6, 21),  date(2018,  9, 22))),
           ('autumn', (date(2018,  9, 23),  date(2018, 12, 21))),
           ('winter', (date(2018, 12, 22),  date(2018, 12, 31)))]
seasons=[]
#Zuordnung der Jahreszeit zum Datum
for i in range(0,dates1.size):
    if dates1asdates[i].year==2015:
        if seasons2015[0][1][0]<=dates1asdates[i]<=seasons2015[0][1][1]:
            seasons.append(seasons2015[0][0])
        if seasons2015[1][1][0]<=dates1asdates[i]<=seasons2015[1][1][1]:
            seasons.append(seasons2015[1][0])
        if seasons2015[2][1][0]<=dates1asdates[i]<=seasons2015[2][1][1]:
            seasons.append(seasons2015[2][0])
        if seasons2015[3][1][0]<=dates1asdates[i]<=seasons2015[3][1][1]:
            seasons.append(seasons2015[3][0])
        if seasons2015[4][1][0]<=dates1asdates[i]<=seasons2015[4][1][1]:
            seasons.append(seasons2015[4][0])
    if dates1asdates[i].year==2016:
        if seasons2016[0][1][0]<=dates1asdates[i]<=seasons2016[0][1][1]:
            seasons.append(seasons2016[0][0])
        if seasons2016[1][1][0]<=dates1asdates[i]<=seasons2016[1][1][1]:
            seasons.append(seasons2016[1][0])
        if seasons2016[2][1][0]<=dates1asdates[i]<=seasons2016[2][1][1]:
            seasons.append(seasons2016[2][0])
        if seasons2016[3][1][0]<=dates1asdates[i]<=seasons2016[3][1][1]:
            seasons.append(seasons2016[3][0])
        if seasons2016[4][1][0]<=dates1asdates[i]<=seasons2016[4][1][1]:
            seasons.append(seasons2016[4][0])
    if dates1asdates[i].year==2017:
        if seasons2017[0][1][0]<=dates1asdates[i]<=seasons2017[0][1][1]:
            seasons.append(seasons2017[0][0])
        if seasons2017[1][1][0]<=dates1asdates[i]<=seasons2017[1][1][1]:
            seasons.append(seasons2017[1][0])
        if seasons2017[2][1][0]<=dates1asdates[i]<=seasons2017[2][1][1]:
            seasons.append(seasons2017[2][0])
        if seasons2017[3][1][0]<=dates1asdates[i]<=seasons2017[3][1][1]:
            seasons.append(seasons2017[3][0])
        if seasons2017[4][1][0]<=dates1asdates[i]<=seasons2017[4][1][1]:
            seasons.append(seasons2017[4][0])
    if dates1asdates[i].year==2018:
        if seasons2018[0][1][0]<=dates1asdates[i]<=seasons2018[0][1][1]:
            seasons.append(seasons2018[0][0])
        if seasons2018[1][1][0]<=dates1asdates[i]<=seasons2018[1][1][1]:
            seasons.append(seasons2018[1][0])
        if seasons2018[2][1][0]<=dates1asdates[i]<=seasons2018[2][1][1]:
            seasons.append(seasons2018[2][0])
        if seasons2018[3][1][0]<=dates1asdates[i]<=seasons2018[3][1][1]:
            seasons.append(seasons2018[2][0])
        if seasons2018[4][1][0]<=dates1asdates[i]<=seasons2018[4][1][1]:
            seasons.append(seasons2018[2][0])
print(len(seasons))
avocado_df['seasons']=seasons
```

```python
#get the mean values for the seasons and average prices

wintersubset=avocado_df[avocado_df["seasons"] == "winter"]
pricesforwinter = wintersubset["AveragePrice"].to_numpy()
priceforwinter = np.mean(pricesforwinter)
print(priceforwinter)

summersubset=avocado_df[avocado_df["seasons"] == "summer"]
pricesforsummer = summersubset["AveragePrice"].to_numpy()
priceforsummer = np.mean(pricesforsummer)
print(priceforsummer)

autumnsubset=avocado_df[avocado_df["seasons"] == "autumn"]
pricesforautumn = autumnsubset["AveragePrice"].to_numpy()
priceforautumn = np.mean(pricesforautumn)
print(priceforautumn)

springsubset=avocado_df[avocado_df["seasons"] == "spring"]
pricesforspring = springsubset["AveragePrice"].to_numpy()
priceforspring = np.mean(pricesforspring)
print(priceforspring)

ywerte=[priceforwinter,priceforspring,priceforsummer,priceforautumn]
xwerte=['winter','spring','summer','autumn']

plt.bar(xwerte, ywerte)
plt.xlabel("season")
plt.ylabel("average price")
plt.show()
```
